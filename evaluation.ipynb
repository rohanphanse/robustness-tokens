```python
import pandas as pd

# Extract feature robustness results from `torchattacks.csv` file
csv_paths = [
    "results/robustness/features/dinov2_vitb14_0",
    "results/robustness/features/dinov2_vitb14_1",
    "results/robustness/features/dinov2_vitb14_5",
    "results/robustness/features/dinov2_vitb14_10",
    "results/robustness/features/dinov2_vitb14_20",
    "results/robustness/features/dinov2_vits14_0",
    "results/robustness/features/dinov2_vits14_10",
    "results/robustness/features/dinov2_vitl14_0",
    "results/robustness/features/dinov2_vitl14_10",
]

for csv_path in csv_paths:
    print(csv_path)
    df = pd.read_csv(f"{csv_path}/cossims.csv")
    print(len(df))
    mean = df.iloc[:, 1].mean()
    print("cossims", mean)
    df = pd.read_csv(f"{csv_path}/mses.csv")
    mean = df.iloc[:, 1].mean()
    print("mses", mean)
    print()
```
